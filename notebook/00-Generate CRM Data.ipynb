```python
# Import the random, pandas, numpy, and datetime libraries for later use
import random 
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
# Set the seed for random number generation
random.seed(2023)

gender_opt = ['MALE','FEMALE']
urb_opt = ['URBAN','RURAL']

mpd_raw = pd.read_csv('../data/MPD_sample_synthetic.csv')
mpd_crm = mpd_raw[['msisdn']].drop_duplicates().reset_index(drop=True)
mpd_crm['gender'] = mpd_crm['msisdn'].apply(lambda x: random.choice(gender_opt))
mpd_crm['age'] = mpd_crm['msisdn'].apply(lambda x: random.randint(6,99))
display(mpd_crm)

mpd_urb = mpd_raw[['cell_id']].drop_duplicates().reset_index(drop=True)
mpd_urb['zone_classification'] = mpd_urb['cell_id'].apply(lambda x: random.choice(urb_opt))
display(mpd_urb)
```

```python
mpd_crm.to_csv('../data/MPD_sample_synthetic_CRM.csv',index=False,sep='\t')
mpd_urb.to_csv('../data/MPD_sample_synthetic_zone.csv',index=False,sep="\t")
```
